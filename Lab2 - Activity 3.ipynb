### Hypothesis 1:  People who do not own smartphones and are less than 50 years play less games

```r
# Selecting the data required to prove the hypothesis

data1 <- survey[c('psraid', 'age','smart1', 'game1')]
```

```r
# Processing the data for visualizing it better

data1 <- subset(data1, smart1 != "8" & smart1 != "9" & game1 != "8" & game1 != "9" )
```

```r
# Saving the dataframe as CSV

my.df <- data.frame(lapply(data1, as.character), stringsAsFactors=FALSE)
write.csv(my.df, file = "hypothesis1.csv")
```

```r
# Setting the labels for all the factors - date1a and emplnw

train <- within(data1, smart1 <- factor(smart1, labels = c("Yes", "No")))
train <- within(train, game1 <- factor(game1, labels = c("Yes", "No")))
```

```r
# Plotting a bar graph to visualize the data 

qplot(age, game1, colour = smart1, data=train, geom="jitter") + labs(title = "SmartPhone Usage to Play Games", y = "Play Video Games", x = "Age")
```

### Hypothesis 2: People employed full-time use the dating apps the least.

```r
# Selecting the data required to prove the hypothesis

data2 <- survey[c('psraid', 'date1a', 'emplnw')]
```

```r
# Processing the data for visualizing it better
data2 <- subset(data2, date1a != "NA" & date1a != "8" & date1a != "9" & emplnw != "98" & emplnw != "99")

# Group by Employment Types and if a person has used a dating app
temp <- data2 %>% group_by(emplnw, date1a)

# Summarising the data to get the counts of people falling in each possible category
temp <- summarise(temp, n_emplnw = length(unique(psraid)))
```

```r
# Saving the dataframe as CSV

my.df <- data.frame(lapply(data2, as.character), stringsAsFactors=FALSE)
write.csv(my.df, file = "hypothesis2.csv")
```

```r
# Setting the labels for all the factors - date1a and emplnw

train <- within(temp, date1a <- factor(date1a, labels = c("Yes", "No")))
train <- within(train, emplnw <- factor(emplnw, labels = c("Full Time", "Part Time", "Retired", 
                                                           "Unemployed", "Self-Employed", "Disabled", 
                                                           "Student", "Other")))
```

```r
# Plotting a bar graph to visualize the data 

par(mfrow = c(6, 7))
ggplot(data = train, aes(x = emplnw, y= n_emplnw, fill=date1a)) + geom_bar(stat = "identity") + labs(title = "Effect of Employment Status on Dating App Usage", y = "No. of People", x = "Employment Status")
```

### Hypothesis 3: People employed in large corporations search less for jobs

```r
# Selecting the data required to prove the hypothesis

data3 <- survey[c('psraid', 'sex', 'emptype1', 'job2')]
```

```r
# Processing the data for visualizing it better
data3 <- subset(data3, emptype1 != "NA" & emptype1 != "98" & emptype1 != "99" & job2 != "8" & job2 != "9" )

# Group by Employment Types and if a person has used a dating app
temp <- data3 %>% group_by(emptype1, sex, job2)

# Summarising the data to get the counts of people falling in each possible category
temp <- summarise(temp, n_emptype = length(unique(psraid)))
```

```r
# Saving the dataframe as CSV

my.df <- data.frame(lapply(data3, as.character), stringsAsFactors=FALSE)
write.csv(my.df, file = "hypothesis3.csv")
```

```r
# Setting the labels for all the factors - date1a and emplnw

train <- within(temp, job2 <- factor(job2, labels = c("Yes", "No")))
train <- within(train, sex <- factor(sex, labels = c("Male", "Female")))
train <- within(train, emptype1 <- factor(emptype1, labels = c("Large Corp.", "Medium Co.", "Small Bus.", 
                                                           "Gov.", "School", "NGO", 
                                                           "Other", "Self-Employed")))
```

```r
# Plotting a bar graph to visualize the data 

par(mfrow = c(6, 7))
ggplot(data = train, aes(x = emptype1, y= n_emptype, fill=job2)) + geom_bar(stat = "identity", position = "dodge") + labs(title = "Relation between Place of Work and Job Search", y = "No. of People", x = "Place of Work")
```

### Hypothesis 4: People with income more than $150,000 are generally married

```r
# Selecting the data required to prove the hypothesis

data4 <- survey[c('psraid', 'age', 'inc', 'marital')]
```

```r
# Processing the data for visualizing it better
data4 <- subset(data4, inc != "NA" & inc != "98" & inc != "99" & marital != "8" & marital != "9" )

# Group by Employment Types and if a person has used a dating app
temp <- data4 %>% group_by(marital, inc, age)

# Summarising the data to get the counts of people falling in each possible category
temp <- summarise(temp, n_inc = length(unique(psraid)))
```

```r
# Saving the dataframe as CSV

my.df <- data.frame(lapply(data4, as.character), stringsAsFactors=FALSE)
write.csv(my.df, file = "hypothesis4.csv")
```

```r
# Setting the labels for all the factors - date1a and emplnw

train <- within(temp, inc <- factor(inc, labels = c("Less than $10,0000", "10 to under $20,000", "20 to under $30,000",
                                                   "30 to under $40,000", "40 to under $50,000", "50 to under $75,000",
                                                   "75 to under $100,000", "100 to under $150,000", "$150,000 or more")))
train <- within(train, marital <- factor(marital, labels = c("Married", "Live In", "Divorced", 
                                                           "Separated", "Widowed", "Never Married")))
```

```r
# Plotting a bar graph to visualize the data 

ggplot(data = train, aes(x = marital, y= inc, color = inc, fill=n_inc)) + geom_jitter(stat = "identity") + labs(title = "Relation between Martial Status and Income", y = "Family Income", x = "Marital Status")
```

### Hypothesis 5: People who do not use internet generally search jobs via Personal connections with close friends or family member

### Hypothesis 6: State California has most educated people.

```r
# Selecting the data required to prove the hypothesis

data6 <- survey[c('psraid', 'educ2', 'state', 'zipcode')]
```

```r
# Processing the data for visualizing it better
data6 <- subset(data6, educ2 != "NA" & educ2 != "98" & educ2 != "99")

# Group by States
temp <- data6 %>% group_by(state)

# Summarising the data to get the counts educated people in each state
temp <- summarise(temp, n_edu = length(unique(psraid)))
```

```r
# Derive top 5 educated states
top_states <- head(temp[order(-temp$n_edu),], 5)
top_states <- data.frame(top_states$state, top_states$n_edu)
names(top_states) <- c("state", "n_edu")

top_states

# Googled the state names for the top 5 states using the zipcode column: 
# states -> [ 6 - California, 48 - Texas, 12 - Florida, 39 - Ohio, 36 - New York]
```

```r
# Saving the dataframe as CSV

my.df <- data.frame(lapply(data6, as.character), stringsAsFactors=FALSE)
write.csv(my.df, file = "hypothesis6.csv")
```

```r
# Setting the labels for all the factors - state

train <- within(top_states, state <- factor(state, labels = c("California", "Florida", 
                                                              "New York", "Ohio","Texas")))
```

```r
# Plotting a bar graph to visualize the data 

ggplot(train, aes(x="", y=n_edu, fill=state))+ geom_bar(width = 1, stat = "identity") + coord_polar("y", start=0) + labs(title = "State-wise Education Level", y = "Number of People")
```
