### Questions on player.csv

```r
players <- read.csv("player.csv") 
```

```r
players
```

#### Which are the top 10 players by overall rating?

```r
rating <- head(players[order(-players$overall_rating),], 10)
rating_ans <- data.frame(rating$player_name, rating$overall_rating)
rating_ans
```

#### Which are the top 10 goalkeepers by sum of gk attributes?

```r
goalkeeper <- head(players[order(-players$gk_total),], 10)
goalkeeper_ans <- data.frame(goalkeeper$player_name, goalkeeper$gk_total)
goalkeeper_ans
```

#### Which are the top 10 players by number of appearances (with any team they played with)?

```r
appearances <- head(players[order(-players$n_appearances),], 10)
appearances_ans <- data.frame(appearances$player_name, appearances$n_appearances)
appearances_ans
```

#### Which are the top 10 players by number of leagues they played in?

```r
leagues <- head(players[order(-players$n_leagues),], 10)
leagues_ans <- data.frame(leagues$player_name, leagues$n_leagues)
leagues_ans
```
